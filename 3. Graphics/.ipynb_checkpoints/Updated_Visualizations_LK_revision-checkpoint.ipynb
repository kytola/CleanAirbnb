```python
#Import libraries
import os
import csv
import matplotlib.pyplot as plt
from matplotlib import dates
import numpy as np

import pandas as pd
import statsmodels.api as sm
import operator
import datetime
```

```python
#Directory setup
cwd1 = os.getcwd()
os.chdir('..')
cwd2 = os.getcwd()
data_path = cwd2 + '/Saved data'
```

```python
#Load in data
os.chdir(data_path)
Portland_data = pd.read_csv('POR_cal_rev_list_FULLMERGE_10_days.csv.gz')
Portland_booking = pd.read_csv('POR_rejoined_booked_df.csv.gz')
Portland_Rev = pd.read_csv('POR_Revs.csv.gz')
```

```python
count_port = Portland_data.groupby(['res_date'])['booked'].sum().rolling(window=30).mean()
```

# Bookings

```python
fig, ax=plt.subplots(1, 1, figsize=(15, 7))
ax.xaxis.set_major_locator(dates.YearLocator())
plt.plot(count_port, '-', linewidth=1, label='Portland')
plt.legend(loc="upper left")
plt.axvline(x='2018-01-16', c='k', linestyle="--", linewidth=.5, label='SF Regulation', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
ax.tick_params(axis='x',rotation=90)
plt.title('Bookings Count Over Time')
plt.xlabel('Date')
plt.ylabel('Bookings')
plt.show()
```

# Listings

```python
list_count_port = Portland_data.groupby(['res_date'])['seen_avail'].sum().rolling(window=30).mean()
```

```python
fig, ax=plt.subplots(1, 1, figsize=(10, 10))
ax.xaxis.set_major_locator(dates.YearLocator())
plt.plot(list_count_port, '-', linewidth=1, label="Portland")
plt.axvline(x='2018-01-16', c='k', linestyle="--", linewidth=.5, label='SF Regulation', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
ax.tick_params(axis='x',rotation=90)
plt.title('Listings Count Over Time')
plt.xlabel('Date')
plt.ylabel('Listings')
plt.show()
```

# Bookings by Host Listings

```python
book_count_port_single = Portland_data[Portland_data['host_listings_count']==1].groupby(['res_date'])['booked'].sum().rolling(window=30).mean()
book_count_port_multi = Portland_data[Portland_data['host_listings_count']>1].groupby(['res_date'])['booked'].sum().rolling(window=30).mean()
```

```python
fix, ax = plt.subplots(figsize=(10,10))
ax.xaxis.set_major_locator(dates.YearLocator())
plt.plot(book_count_port_single, '-', linewidth=1, label="Portland Single")
plt.plot(book_count_port_multi,'-', linewidth=1, label="Portland Multi")
plt.axvline(x='2018-01-16', c='k', linestyle="--", linewidth=1, label='SF Regulation', alpha=0.5)
plt.title('Bookings by Host Listings Number')
plt.xlabel('Date')
plt.ylabel('Bookings')
plt.legend()
plt.show()
```

# Nightly Prices

## Overall

```python
Port_prices = Portland_data.groupby(['res_date'])['calendar_price'].mean()

fix, ax = plt.subplots(figsize=(15,7))
ax.xaxis.set_major_locator(dates.YearLocator())
plt.plot(Port_prices, '--o', linewidth=1, label="Portland")
plt.axvline(x='2018-01-16', c='k', linestyle="--", linewidth=1, label='SF Regulation', alpha=0.5)
plt.title('Nightly Calendar Price Over Time')
plt.xlabel('Date')
plt.ylabel('Calendar Price')
plt.legend()
plt.show()
```
